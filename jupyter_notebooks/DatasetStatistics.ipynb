```python
import os
import pandas as pd
```

```python
dataset_dir = "/home/ghandour/Dataset/ShapeNet_PCD/"
print("Number of classes ", len([name for name in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, name))]))
class_dir = os.path.join(dataset_dir, "airplane")
print("Number of Items in Airplane ", len([name for name in os.listdir(class_dir) if os.path.isfile(os.path.join(class_dir, name))]))
```

```python
dict = {}
for name in os.listdir(dataset_dir):
    dict[name]=0
    class_dir = os.path.join(dataset_dir, name)
    for ply in os.listdir(class_dir):
        dict[name]+=1

print(dict)
```

```python
df = pd.DataFrame.from_dict(dict, columns=["count"], orient='index')
df.sort_values(by="count")
```
